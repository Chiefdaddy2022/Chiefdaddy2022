```python
# Create a figure and specify the size

fig = plt.figure(figsize=(14, 6))  # Specify width and height in inches

# Create a histogram of the 'Sales' column
plt.hist(df['sales'])
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create a figure and specify the size

fig = plt.figure(figsize=(15, 6))  # Specify width and height in inches

# Visualize using a scatter plot

sns.scatterplot(x = df['customerName'], y = df['sales'])
plt.title("Company vs Sales")
plt.xticks(rotation = 90)
plt.show()
```

```python
# Create a figure and specify the size

fig = plt.figure(figsize=(14, 4))  # Specify width and height in inches

# Create the box plot:

sns.boxplot(x='sales', data=df)
plt.title('Box Plot of Sales')
plt.show()
```

```python
# Create a figure and specify the size

fig = plt.figure(figsize=(14, 6))  # Specify width and height in inches

# Create a correlation matrix of multiple continuous variables ('quantity', 'discount', and 'sales')

corr = df[['quantity', 'discount', 'sales']].corr()

# Plot the correlation matrix as a heatmap

sns.heatmap(corr, annot=True, cmap='BuGn_r')
plt.title('Correlation Matrix')
plt.show()
```
